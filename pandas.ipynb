# series

```python
#creating a column
x=[1,2,3,4,5,6]
s=pd.Series(x)
s
```

```python
#creaing a column with index
x=[1,2,3,4,5,6]
s=pd.Series(x,index=['a','b','c','d','e','f'])
s
```

```python
x={'name':'manohar','age':21,'sex':'M'}
s=pd.Series(x)
s
```

# DataFrame

```python
#dataframe is a 2 dimensional array containg rows and columns
x={'names':['manu','sai','lakshmi','mahi'],'age':[21,22,43,20]}
s=pd.DataFrame(x)
s
```

```python
x={'names':['manu','sai','lakshmi','mahi'],'age':[21,22,43,20]}
s=pd.DataFrame(x,index=['o','two','three','four'])
s
```

```python
s.iloc[0]
```

```python
s.iloc[0:2]
```
