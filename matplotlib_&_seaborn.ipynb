```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('iris.csv')
df.drop(['Id'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['species']=label.fit_transform(df['Species'])
```

```python
#bivariate analysis
plt.figure(figsize=(5,3))
sns.lineplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'])
df.drop(['Species'],axis=1,inplace=True)
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(5,3))
sns.lineplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'],ci=None)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'])
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'],hue=df['S'])
```

```python
plt.figure(figsize=(5,3))
sns.histplot(x=df['PetalLengthCm'],kde=True)
```

```python
plt.figure(figsize=(5,3))
sns.distplot(x=df['PetalLengthCm'])
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x=df['PetalLengthCm'])
```

```python
#SUBPLOT
plt.figure(figsize=(10,10))
plot=1
for i in df:
    if plot<=5:
        ax=plt.subplot(3,3,plot)
        sns.distplot(x=df[i])
        plt.xlabel(i)
    plot+=1
plt.tight_layout()
```

```python
#SUBPLOT
plt.figure(figsize=(10,10))
plot=1
for i in df:
    if plot<=5:
        ax=plt.subplot(3,3,plot)
        sns.boxplot(x=df[i])
        plt.xlabel(i)
    plot+=1
plt.tight_layout()
```
